```python
prevTournamentStats = pd.read_csv('March Madness Data/Historical Data/mdf.csv')
prevTournamentStats = prevTournamentStats.drop(['TEAM.1', 'ROUND', 'SEED'], axis=1)
prevTournamentStats
```

```python
#Adding Kenpom Adjusted Efficiency for each team


kenPomAdjEffTeam1 = [];
kenPomAdjEffTeam2 = [];


for i in range(0,len(tournamentGameByGame)):
    teamRowJacob = tournamentGameByGame[tournamentGameByGame['id'] == i]
    team1Name = teamRowJacob.Team1
    #print(team1Name[i])
    team2Name = teamRowJacob.Team2
    year = teamRowJacob.Year
    #print(year[i])
    team1RowJack = prevTournamentStats[((prevTournamentStats['TEAM'] == team1Name[i]) & (prevTournamentStats['YEAR'] == year[i]))]
    team2RowJack = prevTournamentStats[((prevTournamentStats['TEAM'] == team2Name[i]) & (prevTournamentStats['YEAR'] == year[i]))]
    #print()
    
    #team1RowJack = prevTournamentStats.loc(prevTournamentStats['TEAM'])
    kenPomAdjEffTeam1Val = team1RowJack['KENPOM ADJUSTED EFFICIENCY']
    kenPomAdjEffTeam2Val = team2RowJack['KENPOM ADJUSTED EFFICIENCY']
    if(kenPomAdjEffTeam1Val.empty):
        kenPomAdjEffTeam1.append(None)
    else:
        #print(kenPomAdjEffTeam1Val[team1RowJack.index[0]])
        #print('--------------------')
        kenPomAdjEffTeam1.append(kenPomAdjEffTeam1Val[team1RowJack.index[0]])
        
    if(kenPomAdjEffTeam2Val.empty):
        kenPomAdjEffTeam2.append(None)
    else:
        #print(team2Name[i])
        #print(kenPomAdjEffTeam2Val[team2RowJack.index[0]])
        #print('--------------------')
        kenPomAdjEffTeam2.append(kenPomAdjEffTeam2Val[team2RowJack.index[0]])
    
tournamentGameByGame['KENPOM ADJUSTED EFFICIENCY TEAM1'] = kenPomAdjEffTeam1
tournamentGameByGame['KENPOM ADJUSTED EFFICIENCY TEAM2'] = kenPomAdjEffTeam2

tournamentGameByGame

    
    
```

```python
#Generalizing adding stat columns for all stat columns

statTeam1 = [];
statTeam2 = [];

statColName = 'KENPOM ADJUSTED OFFENSE'


for i in range(0,len(tournamentGameByGame)):
    teamRowJacob = tournamentGameByGame[tournamentGameByGame['id'] == i]
    team1Name = teamRowJacob.Team1
    #print(team1Name[i])
    team2Name = teamRowJacob.Team2
    year = teamRowJacob.Year
    #print(year[i])
    team1RowJack = prevTournamentStats[((prevTournamentStats['TEAM'] == team1Name[i]) & (prevTournamentStats['YEAR'] == year[i]))]
    team2RowJack = prevTournamentStats[((prevTournamentStats['TEAM'] == team2Name[i]) & (prevTournamentStats['YEAR'] == year[i]))]
    #print()
    
    #team1RowJack = prevTournamentStats.loc(prevTournamentStats['TEAM'])
    statTeam1Val = team1RowJack[statColName]
    statTeam2Val = team2RowJack[statColName]
    if(statTeam1Val.empty):
        statTeam1.append(None)
    else:
        #print(kenPomAdjEffTeam1Val[team1RowJack.index[0]])
        #print('--------------------')
        statTeam1.append(statTeam1Val[team1RowJack.index[0]])
        
    if(statTeam2Val.empty):
        statTeam2.append(None)
    else:
        #print(team2Name[i])
        #print(kenPomAdjEffTeam2Val[team2RowJack.index[0]])
        #print('--------------------')
        statTeam2.append(statTeam2Val[team2RowJack.index[0]])
    
tournamentGameByGame[f'{statColName} Team1'] = statTeam1
tournamentGameByGame[f'{statColName} Team2'] = statTeam2

tournamentGameByGame
```
