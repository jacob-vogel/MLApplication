```python
class Model(nn.Module):
    def __init__(self):
        super().__init__()
        self.l1 = nn.Linear(17, 21)
        self.l2 = nn.Linear(21, 5)
        self.l3 = nn.Linear(5,1)
        self.fin= nn.Linear(1,1)

    def forward(self, features):
        x= self.l1(features)
        x= self.l2(x)
        x= self.l3(x)
        x= self.fin(x)
        return x

#now write a loss function and an optimizer
model = Model()
loss_fn = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
def train_model(model, loss_fn, optimizer, x, y, epochs=1000):
    for epoch in range(epochs):
        y_pred = model(x)
        loss = loss_fn(y_pred, y)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        if epoch % 1 == 0:
            print(f"Epoch {epoch}, Loss: {loss.item()}")
    return model
```
