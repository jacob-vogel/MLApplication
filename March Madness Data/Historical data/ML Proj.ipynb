```python
def getTeamAverage(df, year, team):
    overallAverage = df.describe().loc['mean']
    averagedf = df.loc[(df['YEAR'] < year) & (df['TEAM']==team)].mean()

    
    if averagedf.isnull().all():
        return overallAverage.to_frame()
    else:
        return averagedf.to_frame()
```
